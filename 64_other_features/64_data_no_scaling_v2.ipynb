# RGB Data

```python
#reading numpy array data
rgb_test = np.load('rgb_data.npy')
#reading csv file
rgb_df = pd.read_csv('rgb_data.csv')
```

```python
#converting pixels npy file to dataframe
features = pd.DataFrame(rgb_test)
```

```python
#extracting mean,min,max,median from  the data frame and adding it to the features
features['mean'] = rgb_df['mean']
features['min'] = rgb_df['min']
features['max'] = rgb_df['max']
features['median'] = rgb_df['median']
```

```python
#checking any negative numbers as a test
(features<0).sum().sum()
```

```python
#printing rgb_df
print(rgb_df)
```

```python
print(features)
```

```python
features = features.values
```

```python
#assigning target
target = rgb_df['disease'].values
#splitting the data to 80 percent and 20 percent
from sklearn.model_selection import train_test_split
Ftrain,Ftest,Ltrain,Ltest=train_test_split(features,target, test_size=0.20, random_state=24)
```

```python
#importing classification report and confusion matrix
from sklearn.metrics import classification_report, confusion_matrix
```

```python
#importing svm model
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
#scaling the data using standard scaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
#sclaing the data
Ftrain_scaled = scaler.fit_transform(Ftrain)
Ftest_scaled = scaler.fit_transform(Ftest)
```

# SVM(RBF) on RGB data

```python
#assigning the model
svm_rbf = SVC(kernel='rbf')
#fitting the model
svm_rbf.fit(Ftrain_scaled,Ltrain)
#predicting the outpur
y_pred = svm_rbf.predict(Ftest_scaled)
print(accuracy_score(Ltest,y_pred))
```

```python
#classification report
y_predict = svm_rbf.predict(Ftest_scaled)
print(classification_report(Ltest, y_predict))
print(confusion_matrix(Ltest, y_predict))
```
