```python
import numpy as np #importing numpy
import pandas as pd #importing pandas
import PIL #importing pillow for image processing
from PIL import Image#importing image from pillow library
import skimage#importing skimage library
import os#importing os library
import time#importing time library

#Capture start time of the program
start_time = time.time()
```

```python
from string import digits
#code to remove numbers in a string and assigning it to the variable
r_digits = str.maketrans('', '', digits)
#taking source data folder path to a varible
rawdata_path = 'D://DKIT//dissertation//Dataset_of_Tomato Leaves//plantvillage//Preprocessed_data//Preprocessed_data//'
#changing current librbay to source data
os.chdir(rawdata_path)
#taking sub-directories to a list
dirs = os.listdir()
directories = []
diseases = []
#Taking directories into a list and extract disease name from folder name
for x in dirs:
    res = x.translate(r_digits).lower()
    diseases.append(res)
    temp = rawdata_path + x + '//'
    directories.append(temp)
```

```python
vector_data = pd.DataFrame(columns=['img_name','gray','gray_single','disease'])
```

```python
x = 0
y=0
for i in directories:
    os.chdir(i)
    k = diseases[x]
    pics = os.listdir()
    for j in pics:
        gray_data = []
        gray_single = []
        gray_l = []
        gray_data.append(j)
        image = Image.open(j,mode='r')
        gray_l = image.convert('L')
        gray_l = np.array(gray_l)
        gray_data.append(gray_l)
        gray_single = gray_l.flatten()
        gray_single = np.array(gray_single)
        gray_data.append(gray_single)
        gray_data.append(k)
        vector_data.loc[len(vector_data)] = gray_data
    x = x+1
```
